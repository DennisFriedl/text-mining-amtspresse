```python
# Wir fügen erst die einzelnen Ausgaben zusammen, um zu verhindern, dass z.B. eine Artikelanzahl von 20 eine 10 mal höhere Repräsentation hat als eine Artikelanzahl von 2
df_PC_ausgabeMean = df_PC.groupby(["Jahr_Monat_Tag"]).mean()
df_NM_ausgabeMean = df_NM.groupby(["Jahr_Monat_Tag"]).mean()

# Zählt die verschiedenen Anzahlen der Artikel
var_cross_tab_PC = pd.crosstab(index=df_PC_ausgabeMean.Artikelanzahl_in_der_Ausgabe, columns="Anzahl")
var_cross_tab_NM = pd.crosstab(index=df_NM_ausgabeMean.Artikelanzahl_in_der_Ausgabe, columns="Anzahl")

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(13,7))
plt.subplots_adjust(top = 0.99, bottom=0.01, hspace=0.4, wspace=0.4) # Abstand zwischen den Graphen

var_cross_tab_PC.plot(kind='bar',
                  ax=ax[0],
                  title='Anzahl der Artikel pro Ausgabe (PC)',
                  )
var_cross_tab_NM.plot(kind='bar',
                  ax=ax[1],
                  title='Anzahl der Artikel pro Ausgabe (NM)',
                  )

ax[0].get_legend().remove() #entfernt die Legende
ax[1].get_legend().remove()

plt.savefig('Anzahl_der_Artikel_pro_Ausgabe.png', bbox_inches='tight')
```

```python
df_PC_ausgabeMeanJahrMean = df_PC_ausgabeMean.groupby(["Jahr"]).mean()
df_NM_ausgabeMeanJahrMean = df_NM_ausgabeMean.groupby(["Jahr"]).mean()

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(13,7))
plt.xticks(range(1863, 1895))

df_PC_ausgabeMeanJahrMean.Woerteranzahl_Ausgabe.plot(kind='line',
                  ax=ax,
                  title='Durchschnittliche Ausgabenlänge in Wörtern',
                  ylim=[1000,6000],
                  label="PC"
                  )
df_NM_ausgabeMeanJahrMean.Woerteranzahl_Ausgabe.plot(kind='line',
                  ax=ax,
                  ylim=[1000,6000],
                  label="NM"
                  )
plt.legend(loc="upper left")
plt.xticks(rotation=70)

plt.savefig('Durchschnittliche_Ausgabenlänge_in_Wörtern.png', bbox_inches='tight')
```

```python
df_PC_Ueberschrift_Ausgabe = df_PC.groupby(["Jahr_Monat_Tag", "Ueberschrift", "Jahr_Monat", "Jahr"]).mean().reset_index() # Legt die Daten nach Ausgabe und Überschrift zusammen.

df_PC_Ueberschrift_Ausgabe.head(20)
```

```python
df_PC_ueberschriftenCount = df_PC_Ueberschrift_Ausgabe.groupby(["Ueberschrift"]).size().reset_index(name='counts') # legt die einzigartigen Überschriften zusammen und zählt diese
df_PC_ueberschriftenCount = df_PC_ueberschriftenCount.set_index(['counts']) # Counts als Index
df_PC_ueberschriftenCount = df_PC_ueberschriftenCount.sort_index(ascending=False) # Absteigend ordnen
df_PC_ueberschriftenCount.head(5)
```

```python
# Wir erstellen uns ein extra Dataframe, das wir mit Nullen füllen und später mit den Dataframes der Überschriften kreuzen.
# Das müssen wir so machen, da .size() bei den Überschriften nur die Vorkommen über 0 behält, wir jedoch später die volle Spanne der Jahre abbilden wollen.
dfZeros = df_NM.groupby(['Jahr_Monat']).size().reset_index(name='counts')
dfZeros = df_PC.groupby(['Jahr_Monat']).size().reset_index(name='counts')
dfZeros = dfZeros.set_index(['Jahr_Monat'])
dfZeros["counts"].values[:] = 0

# Wochenschau
df_PC_Wochenschau = df_PC_Ueberschrift_Ausgabe.loc[df_PC_Ueberschrift_Ausgabe['Ueberschrift'].isin(["Wochenschau.", "Wochenschau"])] # Behält nur die Einträge mit der jeweiligen Überschrift
df_PC_Wochenschau = df_PC_Wochenschau.groupby(['Jahr_Monat']).size().reset_index(name='counts') # Nach Monat gruppieren und zusammengefügte Elemente zählen
df_PC_Wochenschau = df_PC_Wochenschau.set_index(['Jahr_Monat'])
df_PC_Wochenschau = df_PC_Wochenschau.sort_index()
df_PC_Wochenschau = df_PC_Wochenschau.combine_first(dfZeros) # Kreuzung mit der Nuller-Tabelle (siehe oben)

# Vom Landtage
df_PC_Landtage = df_PC_Ueberschrift_Ausgabe.loc[df_PC_Ueberschrift_Ausgabe['Ueberschrift'].isin(["Vom Landtage.", "Vom Landtage"])]
df_PC_Landtage = df_PC_Landtage.groupby(['Jahr_Monat']).size().reset_index(name='counts')
df_PC_Landtage = df_PC_Landtage.set_index(['Jahr_Monat'])
df_PC_Landtage = df_PC_Landtage.sort_index()
df_PC_Landtage = df_PC_Landtage.combine_first(dfZeros)

# Erinnerungen aus dem Vorjahre
df_PC_Vorjahre = df_PC_Ueberschrift_Ausgabe.loc[df_PC_Ueberschrift_Ausgabe['Ueberschrift'].isin(["Erinnerungen aus dem Vorjahre.", "Erinnerungen aus dem Vorjahre"])]
df_PC_Vorjahre = df_PC_Vorjahre.groupby(['Jahr_Monat']).size().reset_index(name='counts')
df_PC_Vorjahre = df_PC_Vorjahre.set_index(['Jahr_Monat'])
df_PC_Vorjahre = df_PC_Vorjahre.sort_index()
df_PC_Vorjahre = df_PC_Vorjahre.combine_first(dfZeros)

# Der Reichstag
df_PC_Reichstag = df_PC_Ueberschrift_Ausgabe.loc[df_PC_Ueberschrift_Ausgabe['Ueberschrift'].isin(["Der Reichstag.", "Der Reichstag"])]
df_PC_Reichstag = df_PC_Reichstag.groupby(['Jahr_Monat']).size().reset_index(name='counts')
df_PC_Reichstag = df_PC_Reichstag.set_index(['Jahr_Monat'])
df_PC_Reichstag = df_PC_Reichstag.sort_index()
df_PC_Reichstag = df_PC_Reichstag.combine_first(dfZeros)

# Vom russisch-türkischen Kriegsschauplatz
df_PC_Kriegsschauplatz = df_PC_Ueberschrift_Ausgabe.loc[df_PC_Ueberschrift_Ausgabe['Ueberschrift'].isin(["Vom russisch-türkischen Kriegsschauplatz.", "Vom russisch-türkischen Kriegsschauplatz"])]
df_PC_Kriegsschauplatz = df_PC_Kriegsschauplatz.groupby(['Jahr_Monat']).size().reset_index(name='counts')
df_PC_Kriegsschauplatz = df_PC_Kriegsschauplatz.set_index(['Jahr_Monat'])
df_PC_Kriegsschauplatz = df_PC_Kriegsschauplatz.sort_index()
df_PC_Kriegsschauplatz = df_PC_Kriegsschauplatz.combine_first(dfZeros)


```

```python
# GRAPH:
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(13,7))
plt.xticks(range(len(df_PC_Wochenschau.index)), df_PC_Wochenschau.index) # Gesamte Zeitspanne als Ticks
ax.set_xticks(ax.get_xticks()[::10]) # Nur jeden zehnten Tick anzeigen

df_PC_Wochenschau.counts.plot(kind='line',
                  ax=ax,
                  title='Vorkommen der häufigsten Überschriften (PC)',
                  ylim=[0,8],
                  label="Wochenschau"                                      
                  )
df_PC_Landtage.counts.plot(kind='line',
                  ax=ax,
                  label="Vom Landtage"                                      
                  )
df_PC_Vorjahre.counts.plot(kind='line',
                  ax=ax,
                  label="Erinnerungen aus dem Vorjahre"                                      
                  )
df_PC_Reichstag.counts.plot(kind='line',
                  ax=ax,
                  label="Der Reichstag"                                      
                  )
df_PC_Kriegsschauplatz.counts.plot(kind='line',
                  ax=ax,
                  label="Vom russisch-türkischen Kriegsschauplatz"                                      
                  )

plt.legend(loc="upper left") # Legende
plt.xticks(rotation=70) # Rotation der Ticks auf der x-Achse
plt.savefig('Vorkommen_Ueberschriften_PC.png', bbox_inches='tight')
```

```python
# Wir erstellen uns ein extra Dataframe, das wir mit Nullen füllen und später mit den Dataframes der Überschriften kreuzen.
# Das müssen wir so machen, da .size() bei den Überschriften nur die Vorkommen über 0 behält, wir jedoch später die volle Spanne der Jahre abbilden wollen.
dfZeros = df_NM.groupby(['Jahr_Monat']).size().reset_index(name='counts')
dfZeros = dfZeros.set_index(['Jahr_Monat'])
dfZeros["counts"].values[:] = 0

# Politische Tagesfrage
df_NM_Tagesfragen = df_NM_Ueberschrift_Ausgabe.loc[df_NM_Ueberschrift_Ausgabe['Ueberschrift'].isin(["Politische Tagesfragen.", "Politische Tagesfragen"])] # Behält nur die Einträge mit der jeweiligen Überschrift
df_NM_Tagesfragen = df_NM_Tagesfragen.groupby(['Jahr_Monat']).size().reset_index(name='counts') # Nach Monat gruppieren und zusammengefügte Elemente zählen
df_NM_Tagesfragen = df_NM_Tagesfragen.set_index(['Jahr_Monat'])
df_NM_Tagesfragen = df_NM_Tagesfragen.sort_index()
df_NM_Tagesfragen = df_NM_Tagesfragen.combine_first(dfZeros) # Kreuzung mit der Nuller-Tabelle (siehe oben)

# Personalien
df_NM_Personalien = df_NM_Ueberschrift_Ausgabe.loc[df_NM_Ueberschrift_Ausgabe['Ueberschrift'].isin(["Personalien.", "Personalien"])]
df_NM_Personalien = df_NM_Personalien.groupby(['Jahr_Monat']).size().reset_index(name='counts')
df_NM_Personalien = df_NM_Personalien.set_index(['Jahr_Monat'])
df_NM_Personalien = df_NM_Personalien.sort_index()
df_NM_Personalien = df_NM_Personalien.combine_first(dfZeros)

# Neuigkeiten aus der Verwaltung
df_NM_Verwaltung = df_NM_Ueberschrift_Ausgabe.loc[df_NM_Ueberschrift_Ausgabe['Ueberschrift'].isin(["Neuigkeiten aus der Verwaltung.", "Neuigkeiten aus der Verwaltung"])]
df_NM_Verwaltung = df_NM_Verwaltung.groupby(['Jahr_Monat']).size().reset_index(name='counts')
df_NM_Verwaltung = df_NM_Verwaltung.set_index(['Jahr_Monat'])
df_NM_Verwaltung = df_NM_Verwaltung.sort_index()
df_NM_Verwaltung = df_NM_Verwaltung.combine_first(dfZeros)

# Aus dem Auslande
df_NM_Ausland = df_NM_Ueberschrift_Ausgabe.loc[df_NM_Ueberschrift_Ausgabe['Ueberschrift'].isin(["Aus dem Auslande.", "Aus dem Auslande"])]
df_NM_Ausland = df_NM_Ausland.groupby(['Jahr_Monat']).size().reset_index(name='counts')
df_NM_Ausland = df_NM_Ausland.set_index(['Jahr_Monat'])
df_NM_Ausland = df_NM_Ausland.sort_index()
df_NM_Ausland = df_NM_Ausland.combine_first(dfZeros)

# Aus dem Inlande
df_NM_Inland = df_NM_Ueberschrift_Ausgabe.loc[df_NM_Ueberschrift_Ausgabe['Ueberschrift'].isin(["Aus dem Inlande.", "Aus dem Inlande"])]
df_NM_Inland = df_NM_Inland.groupby(['Jahr_Monat']).size().reset_index(name='counts')
df_NM_Inland = df_NM_Inland.set_index(['Jahr_Monat'])
df_NM_Inland = df_NM_Inland.sort_index()
df_NM_Inland = df_NM_Inland.combine_first(dfZeros)
```
